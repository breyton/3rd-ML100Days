## 練習時間
假設我們資料中類別的數量並不均衡，在評估準確率時可能會有所偏頗，試著切分出 y_test 中，0 類別與 1 類別的數量是一樣的 (亦即 y_test 的類別是均衡的)

```python
import numpy as np
X = np.arange(1000).reshape(200, 5)
y = np.zeros(200)
y[:40] = 1
```

```python
list_df= list(zip(X, y))
len(list_df)
```

可以看見 y 類別中，有 160 個 類別 0，40 個 類別 1 ，請試著使用 train_test_split 函數，切分出 y_test 中能各有 10 筆類別 0 與 10 筆類別 1 。(HINT: 參考函數中的 test_size，可針對不同類別各自作切分後再合併)

```python
import pandas as pd  
df=pd.DataFrame( list_df, columns = ['X', 'y'])
df
```

```python
y_1 = df.loc[df['y'] == 1]
len(y_1)
```

```python
y_1.describe()
```

```python
from sklearn.model_selection import KFold
skf = KFold(n_splits=5)
train_y1=[]
test_y1=[]
for train, test in skf.split(y_1['X'], y_1['y']):                             
    i=0
    print(i)
    print(len(train), len( test))
    print((sum([y_1.iloc[i,1] for i in train]), (sum([y_1.iloc[i,1] for i in test]))))
    train_y1.append([y_1.iloc[i] for i in train])
    test_y1.append([y_1.iloc[i] for i in test])
    i=1+1
```

```python
test_y1[0]
```

```python
y_0 = df.loc[df['y'] == 0].copy()
len(y_0)
```

```python
y_0.describe()
```

```python
from sklearn.model_selection import KFold
skf = KFold(n_splits=5)
i=0
train_y0=[]
test_y0=[]
for train, test in skf.split(y_0['X'], y_0['y']):                             
    train_y0.append([y_0.iloc[i] for i in train])
    test_y0.append([y_0.iloc[i] for i in test])
    i=i+1
```
