```python
# !pip install brewer2mpl
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
import warnings; warnings.filterwarnings(action='once')

large = 22; med = 16; small = 12
params = {'axes.titlesize': large,
          'legend.fontsize': med,
          'figure.figsize': (16, 10),
          'axes.labelsize': med,
          'axes.titlesize': med,
          'xtick.labelsize': med,
          'ytick.labelsize': med,
          'figure.titlesize': large}
plt.rcParams.update(params)
plt.style.use('seaborn-whitegrid')
sns.set_style("white")
%matplotlib inline

# Version
print(mpl.__version__)  #> 3.0.0
print(sns.__version__)  #> 0.9.0

# Draw Plot
plt.figure(figsize=(16,10), dpi= 80)

df=app_train.loc[app_train['NAME_CONTRACT_TYPE'] == 'Cash loans','AMT_INCOME_TOTAL']
print(df.head())

sns.kdeplot(app_train.loc[app_train['NAME_CONTRACT_TYPE'] == 'Cash loans','AMT_INCOME_TOTAL'], shade=True, color="g", label="Cash loans", alpha=.7)
sns.kdeplot(app_train.loc[app_train['NAME_CONTRACT_TYPE'] == 'Revolving loans','AMT_INCOME_TOTAL'], shade=True,  color="deeppink", label="Revolving loans", alpha=.7)

# Decoration
plt.title('Density Plot by Contract Type', fontsize=22)
plt.legend()
plt.show()
```

```python
sns.kdeplot(app_train.loc[app_train['CODE_GENDER'] == 'M','AMT_INCOME_TOTAL'], shade=True,  color="dodgerblue", label="Male", alpha=.2)
sns.kdeplot(app_train.loc[app_train['CODE_GENDER'] == 'F','AMT_INCOME_TOTAL'], shade=True, color="orange", label="Female", alpha=.2)
# Decoration
plt.title('Density Plot by Gender', fontsize=22)
plt.legend()
plt.show()
```
